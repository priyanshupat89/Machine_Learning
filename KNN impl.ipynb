```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
```

```python
df=pd.DataFrame(np.c_[data.data,data.target],columns=[list(data.feature_names)+['target']])
df
```

```python
df
```

```python
X=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2020)
print("shape of Xtrain data",X_train.shape)
print("shape of X test data",X_test.shape)
print("shape of y train data",y_train.shape)
print("shape of y test data",y_test.shape)
```

# KNN classification

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=50)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

# Elbow

```python
from sklearn.model_selection import cross_val_score
accuracy_rate=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,X,y,cv=10)
    accuracy_rate.append(score.mean())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(range(1,40),accuracy_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red')
```
