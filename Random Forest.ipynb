```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
```

```python
data
```

```python
df=pd.DataFrame(np.c_[data.data,data.target],columns=[list(data.feature_names)+['target']])
```

```python
df
```

```python
X=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test , y_train ,y_test=train_test_split(X,y,test_size=0.33,shuffle=True,random_state=42)
print("X shape is ",X_train.shape)
print("X  shape is ",X_test.shape)
print("y shape is ",y_train.shape)
print("y shape is ",y_test.shape)
```

# Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
RFC=RandomForestClassifier(n_estimators=5,criterion='gini')
```

```python
RFC.fit(X_train,y_train)
```

```python
RFC.score(X_test,y_test)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(RFC.estimators_[0],filled=True)
```

```python
y_pre = RFC.predict(X_test)
y_pre
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print('Classification Report:\n')
print(classification_report(y_test,y_pre))
```

```python
s=accuracy_score(y_pre,y_test)
```
