```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
x
```

```python
y
```

# scatter plot

```python
plt.scatter(x,y,c='r', marker='*')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2d')
plt.savefig('test.png')
```

# line plot

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r*--')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2d')
```

# subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sin wave")
plt.plot(x,y,'r--')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
plt.title("cos wave")
plt.plot(x,y,'r--')
```

# bar plot

```python
x=[2,8,10]
y=[10,16,6]
x1=[3,9,11]
y1=[6,15,7]
x2=[1,8,12]
y2=[2,9,11]
plt.bar(x,y)
plt.bar(x1,y1)
plt.bar(x2,y2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph bar')
```

# histrogram

```python
a=np.array([22,87,5,45,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("histogram")
```
